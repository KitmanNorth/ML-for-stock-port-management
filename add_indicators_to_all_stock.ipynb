```python
# SMA5
def SMA5_cal(df):
    for i in range(0,len(df['Last']) - 4):
        tmp = df.loc[i:i+4,'Last'].mean()
        df.loc[i+4,'SMA5'] = tmp
    for i in range(0,4):
        df.loc[i,'SMA5'] = df.loc[0:i, 'Last'].mean()
            
# SMA12
def SMA12_cal(df):
    for i in range(0,len(df['Last']) - 11):
        tmp = df.loc[i:i+11,'Last'].mean()
        df.loc[i+11,'SMA12'] = tmp
        
# SMA26
def SMA26_cal(df):    
    for i in range(0,len(df['Last']) - 25):
        tmp = df.loc[i:i+25,'Last'].mean()
        df.loc[i+25,'SMA26'] = tmp

# MACD /wo 0-24
def MACD_cal(df):
    k12 = 2/(12 + 1)
    k26 = 2/(26 + 1)
    for i in range(25,len(df)):
        Last = df.loc[i,'Last']
        SMA12 = df.loc[i,'SMA12']
        SMA26 = df.loc[i,'SMA26']
        EMA12 = (Last * k12) + (SMA12 * (1 - k12))
        EMA26 = (Last * k26) + (SMA26 * (1 - k26))
        MACD = EMA12 - EMA26
        df.loc[i,'MACD'] = MACD

# MACD5
def MACD5_cal(df):
    for i in range(0,len(df) - 4):
        tmp = df.loc[i:i+4,'MACD'].mean()
        df.loc[i+4,'MACD5'] = tmp

# Vol5
def Vol5_cal(df):
    for i in range(0,len(df['Volumn']) - 4):
        tmp = df.loc[i:i+4,'Volumn'].mean()
        df.loc[i+4,'Vol5'] = tmp
    for i in range(0,4):
        df.loc[i,'Vol5'] = df.loc[0:i, 'Volumn'].mean()
        
# RSI
def RSI_cal(df):
    for base in range(0,len(df) - 13):
        gain = 0
        loss = 0
        for i in range(0,14):
            if(df.loc[base + i,'ChangePercent'] > 0):
                gain = gain + 1
            else:
                loss = loss + 1
        gain = gain/14
        loss = loss/14
        RSI = 100 - (100/ (1 + (gain/loss)))
        df.loc[base + 13, 'RSI'] = RSI

# RSI5
def RSI5_cal(df):
    for i in range(0,len(df) - 4):
        tmp = df.loc[i:i+4,'RSI'].mean()
        df.loc[i+4,'RSI5'] = tmp

# Money5
def Money5_cal(df):
    for i in range(0,len(df['Money']) - 4):
        tmp = df.loc[i:i+4,'Money'].mean()
        df.loc[i+4,'Money5'] = tmp
    for i in range(0,4):
        df.loc[i,'Money5'] = df.loc[0:i, 'Money'].mean()
        
```
