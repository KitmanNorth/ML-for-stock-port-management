```python
import pandas as pd
df = pd.read_csv('Trade.csv')
df.head()
```

```python
#find all companies
comp = pd.read_csv('Company.csv')
comp
```

```python
comp_names = comp['Symbol'].to_numpy()
for i in comp_names:
    if(i == 'COM7'):
        istock = df[(df['Symbol'] == i)]
        path = r'../ML-for-stock-port-management/Individual-Stock/COM-7.csv' 
        istock.to_csv(path, index = None, header=True)
    else:
        istock = df[(df['Symbol'] == i)]
        path = r'../ML-for-stock-port-management/Individual-Stock/'+ i + '.csv' 
        istock.to_csv(path, index = None, header=True)

#can't use COM7 to create file
```
