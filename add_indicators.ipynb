```python
import pandas as pd
df = pd.read_csv('PTT.csv')
```

```python
df.head()
```

```python
# SMA5
def SMA5_cal(df):
    for i in range(0,len(df['Last']) - 4):
        tmp = df.loc[i:i+4,'Last'].mean()
        df.loc[i+4,'SMA5'] = tmp
    for i in range(0,4):
        df.loc[i,'SMA5'] = df.loc[0:i, 'Last'].mean()
            
# SMA12
def SMA12_cal(df):
    for i in range(0,len(df['Last']) - 11):
        tmp = df.loc[i:i+11,'Last'].mean()
        df.loc[i+11,'SMA12'] = tmp
        
# SMA26
def SMA26_cal(df):    
    for i in range(0,len(df['Last']) - 25):
        tmp = df.loc[i:i+25,'Last'].mean()
        df.loc[i+25,'SMA26'] = tmp

# MACD /wo 0-24
def MACD_cal(df):
    k12 = 2/(12 + 1)
    k26 = 2/(26 + 1)
    for i in range(25,len(df)):
        Last = df.loc[i,'Last']
        SMA12 = df.loc[i,'SMA12']
        SMA26 = df.loc[i,'SMA26']
        EMA12 = (Last * k12) + (SMA12 * (1 - k12))
        EMA26 = (Last * k26) + (SMA26 * (1 - k26))
        MACD = EMA12 - EMA26
        df.loc[i,'MACD'] = MACD

# MACD5
def MACD5_cal(df):
    for i in range(0,len(df) - 4):
        tmp = df.loc[i:i+4,'MACD'].mean()
        df.loc[i+4,'MACD5'] = tmp

# MACD10
def MACD10_cal(df):
    for i in range(0,len(df) - 9):
        tmp = df.loc[i:i+9,'MACD'].mean()
        df.loc[i+9,'MACD10'] = tmp        
        
# Vol5
def Vol5_cal(df):
    for i in range(0,len(df['Volumn']) - 4):
        tmp = df.loc[i:i+4,'Volumn'].mean()
        df.loc[i+4,'Vol5'] = tmp
    for i in range(0,4):
        df.loc[i,'Vol5'] = df.loc[0:i, 'Volumn'].mean()
        
# RSI
def RSI_cal(df):
    for base in range(0,len(df) - 13):
        gain = 0
        loss = 0
        for i in range(0,14):
            if(df.loc[base + i,'ChangePercent'] > 0):
                gain = gain + 1
            else:
                loss = loss + 1
        gain = gain/14
        loss = loss/14
        RSI = 100 - (100/ (1 + (gain/loss)))
        df.loc[base + 13, 'RSI'] = RSI

# RSI5
def RSI5_cal(df):
    for i in range(0,len(df) - 4):
        tmp = df.loc[i:i+4,'RSI'].mean()
        df.loc[i+4,'RSI5'] = tmp

# Money5
def Money5_cal(df):
    for i in range(0,len(df['Money']) - 4):
        tmp = df.loc[i:i+4,'Money'].mean()
        df.loc[i+4,'Money5'] = tmp
    for i in range(0,4):
        df.loc[i,'Money5'] = df.loc[0:i, 'Money'].mean()
    
SMA5_cal(df)
SMA12_cal(df)
SMA26_cal(df)
MACD_cal(df)
MACD5_cal(df)
MACD10_cal(df)
Vol5_cal(df)
RSI_cal(df)
RSI5_cal(df)
Money5_cal(df)
```

```python
def label_options(df):
    tmpMoney = None
    tmpPrice = None
    tmpMACD = None
    tmpRSIBuy = None
    tmpRSISell = None
    for i in range(0,len(df)):
        if df.loc[i,'Money'] > df.loc[i,'Money5'] :
            df.loc[i,'StatusMoney'] = 1
            tmpMoney = 1
        else :
            df.loc[i,'StatusMoney'] = 0
            tmpMoney = 0
            
        if df.loc[i,'Last'] > df.loc[i,'SMA5'] :
            df.loc[i,'StatusPrice'] = 1
            tmpPrice = 1
        elif df.loc[i,'Last'] < df.loc[i,'SMA5'] :
            df.loc[i,'StatusPrice'] = -1
            tmpPrice = -1
        else:
            df.loc[i,'StatusPrice'] = 0
            df.loc[i,'StatusMoney'] = 0
            tmpPrice = 0
            tmpMoney = 0

        if df.loc[i,'MACD'] > 0:
            if df.loc[i,'MACD'] > df.loc[i,'MACD10']:
                df.loc[i,'StatusMACD'] = 1
                tmpMACD = 1
            elif df.loc[i,'MACD'] < df.loc[i,'MACD10']:
                df.loc[i,'StatusMACD'] = -1
                tmpMACD = -1
            else :
                df.loc[i,'StatusMACD'] = 0
                df.loc[i,'StatusMoney'] = 0
                tmpMACD = 0
                tmpMoney = 0
        else :
            df.loc[i,'StatusMACD'] = -1
            tmpMACD = -1

        if df.loc[i,'RSI'] < 70 :
            df.loc[i,'StatusRSIBuy'] = 1
            tmpRSIBuy = 1
        else :
            df.loc[i,'StatusRSIBuy'] = 0
            tmpRSIBuy = 0

        if df.loc[i,'RSI'] > 30 :
            df.loc[i,'StatusRSISell'] = -1
            tmpRSISell = -1
        else :
            df.loc[i,'StatusRSISell'] = 0
            tmpRSISell = 0
            
        if tmpMoney == 1 :
            if tmpPrice + tmpMACD + tmpRSIBuy == 3:
                df.loc[i,'Status'] = 1
            elif tmpPrice + tmpMACD + tmpRSISell == -3:
                df.loc[i,'Status'] = -1
            else :
                df.loc[i,'Status'] = 0
        else :
            df.loc[i,'Status'] = 0
            
label_options(df)
```
