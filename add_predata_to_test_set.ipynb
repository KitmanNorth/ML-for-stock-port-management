```python
import os
import glob
import pandas as pd
```

```python
os.chdir("../ML-for-stock-port-management/Individual-Stock-Test")
extension = 'csv'
all_filenames_test = [i for i in glob.glob('*.{}'.format(extension))]
len(all_filenames_test)
```

```python
os.chdir("../Individual-Stock")
extension = 'csv'
all_filenames_train = [i for i in glob.glob('*.{}'.format(extension))]
len(all_filenames_train)
```

```python
# Get match indices
res = []
for i in all_filenames_test:
    try:
        res.append(all_filenames_train.index(i))
#     res = [all_filenames_train.index(i) for i in all_filenames_test]
    except:
        print(i)
```

```python
# res is index list
len(res)
```

```python
all_filenames = []
for i in range(len(all_filenames_train)):
    if i in res:
        all_filenames.append(all_filenames_train[i])
len(all_filenames)
```

```python
# create csv file with header
for i in all_filenames:
    df = pd.read_csv(i)
    df = pd.DataFrame(columns = df.columns)
    path = '../Individual-Stock-Test-Mod/' + i
    df.to_csv(path, index=None)
```

```python
for i in all_filenames:
    # first 25 records
    df = pd.read_csv(i)
    df24 = df.loc[len(df)-25:,:]
    path = '../Individual-Stock-Test-Mod/' + i
    df24.to_csv(path, mode='a', index=None, header=False)
    # test set
    df_test = pd.read_csv('../Individual-Stock-Test/' + i)
    df_test.to_csv(path, mode='a', index=None, header=False)
    print(i)
```
