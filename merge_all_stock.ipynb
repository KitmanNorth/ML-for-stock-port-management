# Train set

```python
import os
import glob
import pandas as pd
os.chdir("../ML-for-stock-port-management/labeled-stock")
```

```python
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
len(all_filenames)
```

```python
# create csv file with header
df = pd.read_csv(all_filenames[0])
df = pd.DataFrame(columns = df.columns)
df.to_csv('../stock.csv', index=None)
```

```python
for i in range(len(all_filenames)):
    df = pd.read_csv(all_filenames[i])
    df25 = df.loc[25:,]
    df25.to_csv('../stock.csv', mode='a', index=None, header=False)
```

```python
df = pd.read_csv('../stock.csv')
df.head()
```
